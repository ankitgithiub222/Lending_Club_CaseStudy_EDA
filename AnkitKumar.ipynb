### Data Manipulation and Standardization: Deriving  attributes from existing columns and modifying columns

```python
loan_main.head(5)
```

```python
loan_main.info()
```

1. Standardization of int_rate column and converting it to float type 
2. Standardization of emp_length column and converting it to integer type
3. Deriving Year and Month from issue_d for convenince of analysis
4. Standardization of other columns to category format for convenince of analysis


```python
#Removing % character from 	int_rate column and converting it to float type
loan_main['int_rate'] = loan_main['int_rate'].apply(lambda x: float(x[:-2]))
```

```python
# Useing Regular expression to extract integer from emp_length column
import re
loan_main['emp_length'] = loan_main.emp_length.apply(lambda y: re.findall('[0-9]+', str(y))[0])
```

```python
#Formating issue_d to Month and Year for convenince
loan_main.issue_d = pd.to_datetime(loan_main.issue_d, format='%b-%y')

# Deriving New column issue_month consists of month data from issue_d 
loan_main['issue_month'] = pd.to_datetime(loan_main['issue_d']).dt.month

# Deriving New column issue_year consists of year data from issue_d 
loan_main['issue_year'] = pd.to_datetime(loan_main['issue_d']).dt.year
```

```python
#Drpping issue_d column as we already derived required month and year in different coumns
loan_main.drop(['issue_d'], axis=1, inplace=True)
```

```python
#Standardizing columns to category format for convenince of analysis

loan_main.grade = loan_main.grade.astype('category')
loan_main.sub_grade = loan_main.sub_grade.astype('category')
loan_main.loan_status = loan_main.loan_status.astype('category')
loan_main.purpose = loan_main.purpose.astype('category')
loan_main.home_ownership = loan_main.home_ownership.astype('category')
loan_main.term = loan_main.term.astype('category')
loan_main.emp_length = loan_main.emp_length.astype('int64')
loan_main.verification_status = loan_main.verification_status.astype('category')
```

```python
loan_main.info()
```

##### We have now 7 Numerical columns and 9 Catagorical coulumns apart from our target column loan_status
*issue month and year is discrete variable*

```python
#removing  Data from loan status which have value Current as its not relavent to our analysis.
loan_main = loan_main[loan_main.loan_status != 'Current']

loan_main.drop(loan_main.index[loan_main['loan_status'] == 'Current'], inplace=True)
```

## Analyzing other column affecting out target column loan status

```python
#Grouping data based on emp length and loan status
loan_experince_percentage = loan_main.groupby(['loan_status', 'emp_length'])['loan_status'].count()

#unstacking the index to count of charged off and Full paid 
loan_experince_percentage = loan_experince_percentage.unstack(level=0)

#reset Index and make emp_length as a coulmn
loan_experince_percentage = loan_experince_percentage.reset_index()
loan_experince_percentage
```

```python
#Creating new column having Charged_off_percentage data
loan_experince_percentage['Charged_off_percentage'] = round((loan_experince_percentage['Charged Off'] / (loan_experince_percentage['Charged Off'] + loan_experince_percentage['Fully Paid'])) *100 )
loan_experince_percentage
```

```python
plt.figure(figsize=(7,7))
sns.barplot(data = loan_experince_percentage, y ='Charged_off_percentage', x='emp_length', palette='RdYlBu_r' )
plt.title('Percentage of charged off application')
#plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.xlabel('Work Experince')
plt.ylabel('Percentage')
plt.xticks(rotation=0)
plt.show()
```

- Loan Applicants having 10+ Year employee length has maximum chance of default around 16 %
